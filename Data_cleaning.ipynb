<h2>Data Types</h2>
<p>
Data has a variety of types.</b>The main types stored in Pandas dataframes are <b>object</b>, <b>float</b>, <b>int</b>, <b>bool</b> and <b>datetime64</b>. In order to better learn about each attribute, it is always good to know the data type of each column. In Pandas:
</p>

```python
df2.dtypes
```

It returns a Series with the data type of each column.

<p>
It is clear to see that the data type of "symboling" and "curb-weight" are <code>int64</code>, "normalized-losses" is <code>object</code>, and "wheel-base" is <code>float64</code>, etc.
</p>
<p>

```python
df2.describe()
```

<p>
This shows the statistical summary of all numeric-typed (int, float) columns.<br>

```python
# describe all the columns in "df" 
df2.describe(include = "all")
```

<p>
